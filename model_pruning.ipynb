```python
x = df.drop('diagnosis',axis=1)
```

```python
y = df[['diagnosis']]
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y,train_size=.75, random_state=42)
```

# Apply Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
```

```python
clf.fit(xtrain,ytrain)
```

```python
pred = clf.predict(xtest) 
pred
```

```python
np.array(ytest)
```

```python
clf.score(xtest,ytest)
```

# Evaluate the ML model

```python
from sklearn.metrics import classification_report , accuracy_score, confusion_matrix, plot_roc_curve
print(classification_report(ytest,pred)) 
```

```python
test_confusion_matrix = confusion_matrix(ytest,pred)
```

```python
test_confusion_matrix
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
clf1 = RandomForestClassifier() # default parameters, 100 Tree = n_estimators=100

clf1.fit(xtrain, ytrain)
clf1.score(xtest, ytest)
```

# Randomized Search CV

```python
# creating param_distributions for random forest
import numpy as np
trees = np.random.randint(15,150,30)
```

```python
trees
```

```python
max_features = ['sqrt','log2']
max_depth = np.random.randint(2,12,50)
criterion = ['gini','entropy']
min_samples_split = [2,3,4]

grids = {
    'n_estimators' : trees,
    'max_features' : max_features,
    'max_depth' : max_depth,
    'criterion' : criterion,
    'min_samples_split' : min_samples_split
}
```

```python
print(grids)
```

```python
from sklearn.model_selection import RandomizedSearchCV
rcv = RandomizedSearchCV(clf1, grids, n_iter=200, cv=3)  #iteration, cross validation
rcv.fit(xtrain, ytrain)
```

```python
rcv.cv_results_
```

```python
tuning_result = pd.DataFrame(rcv.cv_results_)
tuning_result
```

```python
rcv.best_score_
```

```python
rcv.best_params_
```

```python
clf3 = rcv.best_estimator_
```

```python
clf3.fit(xtrain, ytrain) # training again
```

```python
clf3.score(xtest, ytest)
```
